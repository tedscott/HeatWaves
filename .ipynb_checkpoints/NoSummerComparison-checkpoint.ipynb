```python
# gb = ds_f.SummerLength.where(ds_f.SummerLength == 0).groupby("time.year")
# gb.map(np.count_nonzero)
```

```python
yrs = np.arange(1990,2024,1) #[1996,1997,2005,2006] #
yrs
```

```python
%%time

# takes 3.5 min

for yr in yrs:

    
    # FOURIER
    plt.figure(figsize=(10,6))
    # assign axis and def projection to use
    ax = plt.axes(projection=ccrs.Robinson(central_longitude=0))
    
    # add coastlines and grid
    # gridliner help from https://scitools.org.uk/cartopy/docs/v0.13/matplotlib/gridliner.html
    ax.coastlines()
    ax.gridlines(draw_labels=True,linewidth=1, color='red', alpha=0.5, ylocs=mticker.FixedLocator([-70, -23.5, 23.5, 70]), linestyle='--')
    #ax.ylocator = mticker.FixedLocator([-70, -23.5, 0, 23.5, 70])
    
    ds_f.SummerLength.sel(time=str(yr)).where((ds_f.SummerLength == 0) & ((ds_f.lat >= 23.5) | (ds_f.lat <= -23.5))).plot(
        ax=ax,
        transform=ccrs.PlateCarree(), # assign map projection
        #vmin=-0.7,
        #vmax=0.7, 
        #cmap="coolwarm",#"YlOrRd"#cmap="coolwarm"
        cbar_kwargs={'shrink': 0}
    )
    ax.set_title("Fourier Fit for "+str(yr))

    # POLY
    plt.figure(figsize=(10,6))
    # assign axis and def projection to use
    ax1 = plt.axes(projection=ccrs.Robinson(central_longitude=0))
    
    # add coastlines and grid
    ax1.coastlines()
    ax1.gridlines(draw_labels=True,linewidth=1, color='red', alpha=0.5, ylocs=mticker.FixedLocator([-70, -23.5, 23.5, 70]), linestyle='--')
    #ax1.ylocator = mticker.FixedLocator([-70, -23.5, 0, 23.5, 70])
    
    ds_p.SummerLength.sel(time=str(yr)).where((ds_p.SummerLength == 0) & ((ds_p.lat >= 23.5) | (ds_p.lat <= -23.5))).plot(
        ax=ax1,
        transform=ccrs.PlateCarree(), # assign map projection
        #vmin=-0.7,
        #vmax=0.7, 
        #cmap="coolwarm",#"YlOrRd"#cmap="coolwarm"
        cbar_kwargs={'shrink': 0}
    )
    ax1.set_title("Polynomial Fit for "+str(yr))


plt.show()
```
