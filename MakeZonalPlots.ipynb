## Take in a series of years and produce zonal summer length plot of the series

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from cartopy import crs as ccrs, feature as cfeature
import netCDF4
from netCDF4 import Dataset
from datetime import datetime as dt
```

```python
# what years do I have
!ls -l '../../../Data/ERA5-global/Analysis/New/'
```

```python
plot_years = [1990,1995,2000,2005,2010,2015,2020]
```

## Zonal mean summer lengths

```python
# plot mean summer length across all longitudes for each year of interest

# create list of colors
colors_loc = plt.cm.coolwarm(np.linspace(0, 1, len(plot_years)))

# Make a zonal mean figure of the changes
plt.figure(figsize=(12,6))

# loop through the years and add to plot
for idx, year in enumerate(plot_years):
    input_path = '../../../Data/ERA5-global/Analysis/New/'+str(year)+'_summer_stats.nc'
    ds = xr.open_dataset(input_path)
    zonal_mean = ds.SummerLength.mean(dim='lon')
    latitude = ds.lat
    plt.plot(zonal_mean, latitude, 
             c=colors_loc[idx], label=str(year))

plt.axvline(x=0, color="gray", alpha=1, linestyle=":", linewidth=2)
plt.ylim(-95, 95)
plt.yticks(ticks=np.arange(-90,105,15))
plt.xlabel("Summer length (days)")
plt.ylabel("Latitude ($^{\circ}$)")
plt.title(
    str(plot_years[0]) + "-" + str(plot_years[-1]) + " Summer length by latitude (days)",  
    loc="center",
)
plt.legend()
plt.savefig("./Plots/"+str(plot_years[0]) + "-" + str(plot_years[-1])+"_SummerLength_zonal.png")
plt.show()
```
