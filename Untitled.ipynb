```python
import numpy as np
```

```python
x0 = 5
x1 = x0 - (np.log(x0) + x0 - 6)/(1/x0 + 1)
print("x1 = "+str(x1))
x2 = x1 - (np.log(x1) + x1 - 6)/(1/x1 + 1)
print("x2 = "+str(x2))
x3 = x2 - (np.log(x2) + x2 - 6)/(1/x2 + 1)
print("x3 = "+str(x3))
```
