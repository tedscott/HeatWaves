```python
# playing with gradient functions
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,10)
ts = np.array([1,2,4,8,16,8,4,2,1])
grad = np.gradient(ts)
grad, ts, x
```

```python
plt.figure()
plt.plot(x,ts, c='b')
plt.scatter(x,grad, c='r')
```
