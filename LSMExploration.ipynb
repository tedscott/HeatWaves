## Examine the ERA5 Land-Sea mask

#### From the documentation(https://codes.ecmwf.int/grib/param-db/172): This parameter is the proportion of land, as opposed to ocean or inland waters (lakes, reservoirs, rivers and coastal waters), in a grid box.

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from cartopy import crs as ccrs, feature as cfeature
import netCDF4
from netCDF4 import Dataset
from datetime import datetime as dt
```

```python
file_path = '../../../Data/ERA5-global/ERA5-2023-09-01-Full-LSM-download.nc'
ds = xr.open_dataset(file_path)
ds
```

```python
# adjust longitude to run from -180 to 180
# solution from: https://stackoverflow.com/questions/53345442/about-changing-longitude-array-from-0-360-to-180-to-180-with-python-xarray
# save attributes so I can add them back later
lon_atrib = ds.coords['longitude'].attrs  
lat_atrib = ds.coords['latitude'].attrs
ds.coords['longitude'] = (ds.coords['longitude'] + 180) % 360 - 180
ds = ds.sortby(ds.longitude)
ds
```

```python
# also save that out because that is how I will want the longitude
# also rename to lat and lon like the t2m data
ds = ds.rename({'longitude':'lon','latitude':'lat'})
ds = ds.mean(dim='time') # get rid of time dimension since there is only one value
ds.coords['lon'].attrs = lon_atrib
ds.coords['lat'].attrs = lat_atrib
output_path = '../../../Data/ERA5-global/ERA5-2023-09-01-CoordFixed-LSM.nc'
ds.to_netcdf(output_path)
ds
```

```python
plt.figure(figsize=(12,8))
ds.lsm.plot()
```

```python
margins = ds.where((ds.lsm > 0.3) & (ds.lsm < 0.9))
margins = margins.mean(dim='time') # to drop the time dimension since there is only one
plt.figure(figsize=(12,6))
margins.lsm.plot()
```

```python
# make a north america map - lots of stuff getting included in Canada
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(1,1,1, projection=ccrs.PlateCarree())  
ax.set_title("LSM between 0.2 and 0.8 in North America", loc="left")
dataplot = ax.contourf(
    margins.longitude,
    margins.latitude,
    margins.lsm,
    levels=np.arange(0,1,0.1),
    transform=ccrs.PlateCarree(),
    cmap=plt.cm.coolwarm,
)
#ax.coastlines()
ax.gridlines(draw_labels=True, linewidth=1, color='black', alpha=0.5, linestyle='--')
ax.set_extent((-180, -50, 10, 75))
#ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='black')
#ax.add_feature(cfeature.STATES, linewidth=0.3, edgecolor='brown')
fig.colorbar(dataplot, orientation="vertical", label="Proportion of Land",
             shrink=0.55, pad=0.11
            )
#plt.savefig(str(input_year)+"_SummerHeat_NA.png")
plt.show()
```

```python
oceans = ds.where(ds.lsm <= 0.2) 
oceans = oceans.mean(dim='time') # to drop the time dimension since there is only one
plt.figure(figsize=(12,6))
oceans.lsm.plot()
```

```python
land = ds.where(ds.lsm > 0.8) 
land = land.mean(dim='time') # to drop the time dimension since there is only one
plt.figure(figsize=(12,6))
land.lsm.plot()
```
